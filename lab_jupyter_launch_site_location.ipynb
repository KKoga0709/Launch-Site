```python
## Task 1: Mark all launch sites on a map
import folium

# Define the coordinates of the launch site (example: CCAFS LC-40)
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Define the coordinates of the closest coastline (example: coastline near CCAFS LC-40)
coastline_lat = 28.56367
coastline_lon = -80.57163

#KSC location
ksc_lat = 28.573255
ksc_lon = -80.646895

#VAFB location
vafb_lat = 34.632834
vafb_lon = -120.610745

# Create the map centered around the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=14)

# Define the coordinates for the PolyLine (launch site and coastline)
coordinates = [
    [launch_site_lat, launch_site_lon],  # Launch site coordinates
    [coastline_lat, coastline_lon],
    [ksc_lat, ksc_lon],
    [vafb_lat, vafb_lon]# Coastline coordinates
]

# Create a `folium.PolyLine` object
line = folium.PolyLine(locations=coordinates, weight=2, color='blue')

# Add the PolyLine to the map
site_map.add_child(line)

# Add markers for the launch site and coastline
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup='Launch Site (CCAFS LC-40)',
    icon=folium.Icon(color='green')
).add_to(site_map)

folium.Marker(
    location=[coastline_lat, coastline_lon],
    popup='Coastline',
    icon=folium.Icon(color='red')
).add_to(site_map)

folium.Marker(
    location=[ksc_lat, ksc_lon],
    popup='Coastline',
    icon=folium.Icon(color='red')
).add_to(site_map)

folium.Marker(
    location=[vafb_lat, vafb_lon],
    popup='Coastline',
    icon=folium.Icon(color='red')
).add_to(site_map)

# Display the map
site_map.save("launch_map.html")  # Saves the map as an HTML file

# If you're using a Jupyter notebook, just call site_map to display it inline
site_map
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
def marker_color(color):
    if color == 1:
        return 'green'
    else:
        return 'red'
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
spacex_df['marker_color'] = spacex_df['class'].apply(marker_color)
spacex_df.tail(5)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
def marker_color(result):
    return 'green' if result == 1 else 'red'
spacex_df['marker_color'] = spacex_df['class'].apply(marker_color)

from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster()
site_map.add_child(marker_cluster)

for _, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    folium.Marker(
        location=coordinate,
        icon=folium.Icon(color=row['marker_color'], icon_color=row['marker_color'])
    ).add_to(marker_cluster)

site_map.save("launch_outcomes_map.html")
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_coastline
```

```python
# Assuming `line_coordinates` should be a list of pairs like [[lat1, lon1], [lat2, lon2], ...]
line_coordinates = [[row['Lat'], row['Long']] for _, row in spacex_df.iterrows()]

# Iterate over each pair of coordinates
for coordinates in line_coordinates:
    # Calculate distance to coastline
    distance_to_coastline = calculate_distance(launch_site_lat, launch_site_lon, coordinates[0], coordinates[1])
    
    # Adjust marker coordinates as needed
    marker_coordinates = [coordinates[0] + 0.01, coordinates[1] - 0.01]
    
    # Create a distance marker
    distance_marker = folium.Marker(
       marker_coordinates,
       icon=DivIcon(
           icon_size=(20, 20),
           icon_anchor=(0, 0),
           html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_to_coastline),
       )
    )
    site_map.add_child(distance_marker)
    
    # Create a PolyLine for lines
    lines = folium.PolyLine(locations=[(launch_site_lat, launch_site_lon), coordinates], weight=5)
    site_map.add_child(lines)

# Save the map as an HTML file
site_map.save("map_with_markers_and_lines.html")
```
